```python
import requests
from pprint import pprint
from wordcloud import WordCloud
#import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
```

```python
res = requests.get('https://newsapi.org/v2/everything', params = {'q': 'China GDP', 'apiKey':api_key})
```

```python
pprint(res.json())
```

```python
type(res.json())
```

```python
res_json = res.json()
```

```python
all_text = '\n'.join([art['description'] for art in res_json['articles']])
```

```python
print(all_text)
```

```python
# lower max_font_size, change the maximum number of word and lighten the background:
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(all_text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
